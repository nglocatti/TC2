**1.** Hallar la transferencia $T= \frac{V_{o}}{V_{i}}$ en función de $ω_{o}$ y $Q$.

Transferencia hallada:

$ T(S) = \frac{V_{o}}{V_{i}} = -\frac{R_{3}}{R_{1}} . \frac{ \frac{1}{C^2 R_{3}^2}}{S^2 + S \frac{1}{C R_{2}} + \frac{1}{C^2 R_{3}^2}}  $

$ T(S) = \frac{V_{o}}{V_{i}} = K . \frac{ \omega_{o}^2}{S^2 + S \frac{\omega_{o}}{Q} + \omega_{o}^2}  $

$ \omega_{o} = \frac{1}{C R_{3}} $

$ K = -\frac{R_{3}}{R_{1}} $

$ Q = \frac{R_{2}}{R_{3}} $

**2.** Obtener el valor de los componentes del circuito de forma tal que $ω_{o}=1$ y $Q=3$

$ \omega_{o} = 1 = \frac{1}{CR_{3}} $

$ R_{3} = \frac{1}{C} $ 

Si adoptamos $ R_{3} = 10K\Omega $

$ C = 100 uF $

$ Q = 3 = \frac{R_{2}}{R_{3}} $

$ R_{2} = 3 . 10k\Omega $

$ R_{2} = 30k\Omega $

***

**3.** Ajustar el valor de $R_{1}$ de forma tal que $|T(0)|=20dB$.

$ |T(0)|_{dB} = 20 log( |T(0)|_{veces} ) $

$ 1 = log( |T(0)| ) $

$ |T(0)| = 10 = K $

$ K = \frac{R_{3}}{R_{1}} $

$ R_{1} = \frac{R_{3}}{K} = \frac{10k\Omega}{10} $

$ R_{1} = 1K\Omega $
